```python
armies={0: [], 1: []}
dead_units = []
for frame_number in range(len(rep)):
    frame = rep.getFrame(frame_number)
    units = frame.units
    for player_id, unit_arr in units.iteritems():
        # Player -1 is map resources
        if player_id < 0:
            continue
        for unit in unit_arr:
            if unit.id in dead_units:
                # Ignore units that are dead
                continue
            # Get name for comparison and saving
            unit_name = type_to_name(unit.type)
            if unit_name == "Not a valid attacking unit":
                # Army doesn't contain buildings
                continue
            if [unit.id, unit_name] in armies[player_id]:
                # Don't add if it already exists
                continue
            else:
                # Add newcomer to the ranks
                armies[player_id].append([unit.id, unit_name])
                print("Player " + str(player_id) + " gained a " + unit_name)
            if unit.health < unit.max_health and unit.completed:
                dead_units.append(unit.id)
                armies[player_id].remove([unit.id, unit_name])
                print("Player " + str(player_id) + " lost a " + unit_name)
```
